# Problema 1

Sea X = número de águilas al lanzar 3 veces una moneda honesta.

### (a) Espacio muestral

Ω = {AAA, AAS, ASA, ASS, SAA, SAS, SSA, SSS}

### (b) Función de probabilidad

X ~ Binomial(n=3, p=0.5)

P(X = k) = C(3,k)(0.5)^k (0.5)^(3-k)

P(0)=1/8  
P(1)=3/8  
P(2)=3/8  
P(3)=1/8  

### (c) Esperanza

E(X)=np=3(0.5)=1.5

### (d) Varianza

Var(X)=np(1-p)=3(0.5)(0.5)=0.75

**Interpretación:** En promedio se esperan 1.5 águilas.

# Problema 3

```r

datos <- c(500,500,495,490,473,441,429,419,405,400,390,390,390,376,353,
           350,300,258,240,220,210,200,192,190,150,130,125,110,100,100)

anio1 <- datos[1:15]
anio2 <- datos[16:30]

mean(anio1); median(anio1); var(anio1); sd(anio1)
mean(anio2); median(anio2); var(anio2); sd(anio2)

quantile(datos, c(0.16,0.84))
quantile(datos, c(0.1,0.5,0.9))

hist(datos, col="gray")
abline(v=mean(datos), col="red", lwd=2)
abline(v=median(datos), col="blue", lwd=2)
```


**Interpretación:**  
Se observa asimetría positiva.


# Problema 4

```r

library(ggplot2)

sexo <- c(rep("Hombre",20), rep("Mujer",16))
estado <- c(rep("Soltero",10), rep("Casado",10),
            rep("Soltero",8), rep("Casado",8))

tabla <- table(sexo, estado)
tabla

margin.table(tabla,1)
margin.table(tabla,2)

prop.table(tabla,1)
prop.table(tabla,2)

df <- as.data.frame(tabla)

ggplot(df, aes(x=sexo, y=Freq, fill=estado)) +
  geom_bar(stat="identity", position="fill") +
  labs(title="Proporción de Estado Civil por Sexo",
       y="Proporción") +
  scale_fill_grey()
```

```r
install.packages("ggplot2")
```


**Interpretación:**  
Se analizan proporciones marginales y condicionales.
